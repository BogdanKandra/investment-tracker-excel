***Generate a CSV file for importing the investment portfolio saved in the `portfolio.json` file into Yahoo Finance***

### Imports

```python
import subprocess
from pathlib import Path
```

### Set paths

```python
PROJECT_PATH = Path.cwd().parent
GENERATION_SCRIPT_PATH = PROJECT_PATH / "src" / "csv_generators" / "yfinance.py"
```

### Generate the Yahoo Finance import CSV file

```python
commands = ["python", GENERATION_SCRIPT_PATH]

process = subprocess.run(
    commands,
    capture_output=True,
    text=True
)

print(process.stdout)
```
