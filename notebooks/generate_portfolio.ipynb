***Generate an Excel file for managing the investment portfolio saved in the `portfolio.json` file***

### Imports

```python
import subprocess
from pathlib import Path
```

### Set paths and arguments

```python
PROJECT_PATH = Path.cwd().parent
GENERATION_SCRIPT_PATH = PROJECT_PATH / "src" / "generate_investment_portfolio.py"
```

```python
TEST_GENERATION = True
```

### Generate the portfolio management Excel file

```python
commands = ["python", GENERATION_SCRIPT_PATH]
if TEST_GENERATION:
    commands.append("--test")

process = subprocess.run(
    commands,
    capture_output=True,
    text=True
)

print(process.stdout)
```
